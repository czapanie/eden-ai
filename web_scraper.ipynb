```python
from bs4 import BeautifulSoup
import lxml
import requests
```

```python
website = 'https://zszp.pl/szkolka/cyrzan-szkolki-drzew-i-krzewow-ozdobnych/#'
result = requests.get(website)
```

```python
content = result.text
```

```python
content
```

```python
soup = BeautifulSoup(content)
```

```python
soup
```

```python
box = soup.find('div',class_="column_right_60")
```

```python
print(box)
```

```python
text = box.find('a').get_text()
print(text)
```

```python
box.findAll('a')
```

```python
for line in box.findAll('a'):
        text += line.get_text()
        text += '/n '
print(text)
```

```python
with open('data.txt', 'w') as file:
    for line in box.findAll('a'):
        line.get_text()
        file.write(text + "/n")
    
        
```
